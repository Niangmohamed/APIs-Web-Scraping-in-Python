## Pagination

**Pagination** is used to manage queries that return a large number of results. 

Let's determine all the repositories of the **first page starred** on the **Github of Karpathy**.

```python
params = {"per_page" : 50, "page" : 1}
p1_starred_repos_response = requests.get("https://api.github.com/users/karpathy/starred", 
                                         headers=headers, 
                                         params=params).json()
p1_starred_repos_response
```

## User-level access point

The **user-level access point** refers to the fact that the server only keeps track of the user who has logged in with a well-defined token. So once we authenticate, we don't need to specify the name of the user, we can put user directly and the query will interpret user directly as the user who authenticated.

```python
# Example
user_access_point = requests.get("https://api.github.com/user", headers=headers).json() # user=Niangmohamed
user_access_point
```

## POST request

The goal of the **POST** request is to **send informations (create an object)** to the **servers API**. For example, for Github, you can use it to create repositories. Note that for each request, the API is free to accept it or not.

```python
# Example
repos = {"name":"API-Authentication-in-Python"}
add_repos_response = requests.post("https://api.github.com/user/repos", json=repos, headers=headers)
```

```python
add_repos_response_status = add_repos_response.status_code
add_repos_response_status
```

The code **201** shows that the request worked and created the github *API-Authentication-in-Python* repository.

## PATCH / PUT request

The **PATCH** request is used to change attributes of an object. For example changing the name or description of a repository. And we use the **PUT** request when we want to change the object completely. All these requests depend on the **APIs** used. Some **APIs** can accept only the **PUT** request, other **APIs** can accept a **PUT** request as a **PATCH** request.

```python
# Example
repos_edited = {
    "name":"API-Authentication", 
    "description":"Access and request an API and prepare for scraping!"}
modify_repos_response = requests.patch("https://api.github.com/repos/Niangmohamed/API-Authentication-in-Python", 
                                    json=repos_edited, headers=headers)
```

```python
modify_repos_response_status = modify_repos_response.status_code
modify_repos_response_status
```

In github, the **PUT** request works the same way as **PATCH**.

## DELETE request

The **DELETE** request allows to delete an object from an API.

```python
delete_repos_response = requests.delete("https://api.github.com/repos/Niangmohamed/API-Authentication-in-Python", 
                                       headers=headers)
```

```python
delete_repos_response_status = delete_repos_response.status_code
delete_repos_response_status
```
